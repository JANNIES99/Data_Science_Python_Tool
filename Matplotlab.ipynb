```python
import matplotlib.pyplot as plt
import numpy as np
```

# **Simple Plot Graph**

```python
plt.plot([100,200,300,400,500],[1,2,3,4,5])
plt.show()
```

```python
x_vals = [1,2,3,4,5]
y_vals = [1,4,9,16,25]
plt.plot(x_vals, y_vals,label='y=x^2')
plt.ylabel('The y-axis label!')
plt.xlabel('The x-axis label!')
plt.title("The title of the graph!")
plt.xticks(x_vals,['One','Two','Three','Four','Five'])
plt.legend()
plt.show()
```

**Sine And Cosine Curve**

```python
x_vals=np.arange(0,2*np.pi,0.1)
y_vals1=np.sin(x_vals)
y_vals2=np.cos(x_vals)
plt.figure(figsize=(10,10))
plt.plot(x_vals,y_vals1,label='sin(x)',linestyle='dotted',color='r')
plt.plot(x_vals,y_vals2,label='cos(x)',linestyle='dashed',color='b')
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))
fig.subplots_adjust(hspace=0.8, wspace=0.8)
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    ax.text(0.5, 0.5, str((2, 3, i)),
           fontsize=18, ha='center')
```

# **Bar Plot Graph**

```python
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [140, 100, 30, 80]
bar_colors = ['red', 'blue', 'green', 'orange']

plt.bar(fruits, counts, color=bar_colors)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12,7))
fig.subplots_adjust(hspace=0.4)
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

bar_labels = ['R','B','G','Y']


axes[0,0].plot(x, y_sin, label = "Sine")
axes[0,0].plot(x, y_cos, label = "Cosine")
axes[0,0].set_xlabel('x axis label')
axes[0,0].set_ylabel('y axis label')
axes[0,0].set_title('Basic')
axes[0,0].legend()

axes[0,1].bar(fruits, counts, label=bar_labels, color=bar_colors)
axes[0,1].set_title('Bar')
axes[0,1].legend()


axes[1,0].scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
axes[1,0].set_title('Scatter')

axes[1,1].hist(np.random.randn(100), bins=20, color='r', alpha=0.8)
axes[1,1].set_title('Histogram')
```

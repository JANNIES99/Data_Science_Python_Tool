```python
import matplotlib.pyplot as plt
import numpy as np
```

# **Plot Graph**

```python
plt.plot([100,200,300,400,500],[1,2,3,4,5])
plt.show()
```

```python
x_vals = [1,2,3,4,5]
y_vals = [1,4,9,16,25]
plt.plot(x_vals, y_vals,label='y=x^2')
plt.ylabel('The y-axis label!')
plt.xlabel('The x-axis label!')
plt.title("The title of the graph!")
plt.xticks(x_vals,['One','Two','Three','Four','Five'])
plt.legend()
plt.show()
```

**Sine And Cosine Curve**

```python
x_vals=np.arange(0,2*np.pi,0.1)
y_vals1=np.sin(x_vals)
y_vals2=np.cos(x_vals)
plt.figure(figsize=(10,10))
plt.plot(x_vals,y_vals1,label='sin(x)',linestyle='dotted',color='r')
plt.plot(x_vals,y_vals2,label='cos(x)',linestyle='dashed',color='b')
plt.legend()
plt.show()
```
